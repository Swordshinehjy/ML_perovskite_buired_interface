```python
import numpy as np
from rdkit import Chem
from PyFingerprint.fingerprint import get_fingerprint
import pandas as pd
mols = Chem.SmilesMolSupplier('./smiles.txt')
smiles = [Chem.MolToSmiles(i) for i in mols]
```

```python
morgans = [get_fingerprint(i, 'morgan').to_numpy().astype(int) for i in smiles]
datamorgan = np.vstack(morgans)
morgan = pd.DataFrame(datamorgan, columns=[f"morgan{i}" for i in range(1, 2049)])
morgan.head()
```

```python
morgan.to_csv('morgan.csv', index=False)
```

```python
krfps = [get_fingerprint(i, 'klekota-roth').to_numpy().astype(int) for i in smiles]
data_krfp = np.vstack(krfps)
krfp = pd.DataFrame(data_krfp, columns=[f"KRFP{i}" for i in range(1, 4861)])
krfp.head()
```

```python
krfp.to_csv('krfp.csv', index=False)
```

```python
maccss = [get_fingerprint(i, 'maccs').to_numpy().astype(int) for i in smiles]
data_maccs = np.vstack(maccss)
maccs = pd.DataFrame(data_maccs, columns=[f"MACCS{i}" for i in range(1, data_maccs.shape[1]+1)])
maccs.head()
```

```python
maccs.to_csv('maccs.csv', index=False)
```

```python
ecfp6s = [get_fingerprint(i, 'circular').to_numpy().astype(int) for i in smiles]
data_ecfp = np.vstack(ecfp6s)
ecfp = pd.DataFrame(data_ecfp, columns=[f"ECFP{i}" for i in range(1, data_ecfp.shape[1]+1)])
ecfp.head()
```

```python
ecfp.to_csv('ecfp.csv', index=False)
```
