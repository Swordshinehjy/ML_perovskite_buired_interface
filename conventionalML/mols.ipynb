```python
import numpy as np 
import pandas as pd
from pandas import read_csv
from rdkit import Chem
from rdkit.Chem.Draw import MolsToGridImage
#Read data
data = read_csv('dataset.csv')
data.head()
```

```python
mols = [m for m in Chem.SmilesMolSupplier('smiles.txt')]
legends = [f"{control}%->{pce}%" for control, pce in zip(data['control_PCE'], data["PCE"])]
```

```python
MolsToGridImage(mols[0:50],subImgSize=(300,250),molsPerRow=6,legends=legends[0:50])
```

```python
MolsToGridImage(mols[50:100],subImgSize=(300,250),molsPerRow=6,legends=legends[50:100])
```

```python
MolsToGridImage(mols[100:150],subImgSize=(300,250),molsPerRow=6,legends=legends[100:150])
```

```python
MolsToGridImage(mols[150:],subImgSize=(300,250),molsPerRow=6,legends=legends[150:])
```
